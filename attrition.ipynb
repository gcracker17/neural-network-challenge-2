## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()

```

```python
# Create a list of at least 10 column names to use as X data
x_columns = ['Age', 'DistanceFromHome', 'Education', 'JobSatisfaction', 'HourlyRate',
            'OverTime', 'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears', 'YearsAtCompany']


# Create X_df using your selected columns
X_df = attrition_df[x_columns]

# Display the first few rows of X_df
X_df.head()



# Show the data types for X_df
X_df.dtypes


```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

```

```python
# Convert your X data to numeric data types however you see fit
X_df = pd.get_dummies(X_df)
# Add new code cells as necessary
X_df.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_df)


# Scale the training and testing data
X_scaled = X_scaler.transform(X_df)

```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
OHE_Department = OneHotEncoder(categories='auto')

# Fit the encoder to the training data
OHE_Department.fit(y_df[['Department']])


# Create two new variables by applying the encoder
# to the training and testing data
y_encoded = OHE_Department.transform(y_df[['Department']])


```

```python
# Create a OneHotEncoder for the Attrition column
OHE_Attrition = OneHotEncoder(categories='auto')


# Fit the encoder to the training data
OHE_Attrition.fit(y_df[['Attrition']])


# Create two new variables by applying the encoder
# to the training and testing data
y_encoded = OHE_Attrition.transform(y_df[['Attrition']])

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
number_columns_features = len(X_df.columns)
number_columns_features


# Create the input layer
input = layers.Input(shape=(number_columns_features,))


# Create at least two shared layers
shared_layer_1 = layers.Dense(units=64, activation='relu')(input)
shared_layer_2 = layers.Dense(units=32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department




# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_department = layers.Dense(units=32, activation='relu')(shared_layer_2)

# Create the output layer
output_department = layers.Dense(units=2, activation='softmax')(hidden_layer_department)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_attrition = layers.Dense(units=32, activation='relu')(shared_layer_2)

# Create the output layer
output_attrition = layers.Dense(units=2, activation='softmax')(hidden_layer_attrition)

```

```python
# Create the model
model = Model(inputs=input, outputs=[output_department, output_attrition])


# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_scaled, y_encoded, epochs=100, batch_size=32)

```
